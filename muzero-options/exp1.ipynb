```python
import pickle
from MuZero import MuZero
from GridWorldOption import GridWorldOption, GridWorldMacroAction

from GridWorld import GridWorld

from utils import plot_mu_v, plot_trajectory

import numpy as np
```

```python
with open('results/mu_mult.pickle', 'rb') as f:
    mu = pickle.load(f)
    
with open('results/mu_ma_mult.pickle', 'rb') as f:
    mu_ma = pickle.load(f)
    
with open('results/mu_macro_mult.pickle', 'rb') as f:
    mu_macro = pickle.load(f)
```

```python
MAP_NAME = './maps/den204d_hard.map'
with open(MAP_NAME) as f:
    env_map = f.read()
    
env = GridWorld(env_map, 30000)
```

```python
scores_las = []
score_ma_las = []
score_macro_las = []

sims_las = []
sims_ma_las = []
sims_macro_las = []

for i in range(10):
    print(i)
    score_la, sims_la, options_la, states_la = mu.play_simple(env)
    score_ma_la, sims_ma_la, options_ma_la, states_ma_la = mu_ma.play_simple(env)
    score_macro_la, sims_macro_la, options_macro_la, states_macro_la = mu_macro.play_simple(env)
    
    scores_las.append(score_la)
    score_ma_las.append(score_ma_la)
    score_macro_las.append(score_macro_la)
    
    sims_las.append(sims_la)
    sims_ma_las.append(sims_ma_la)
    sims_macro_las.append(sims_macro_la)
```

```python
print('single actions, score:', np.mean(scores_las), 'sims: ', np.mean(sims_la))
print('out of the room option',  np.mean(score_ma_las), 'sims: ', np.mean(sims_ma_la))
print('macro actions',  np.mean(score_macro_las),  np.mean(sims_macro_la))
```
