```python
import graphviz
from MCTS import MCTS
from MountainCarOption import MountainCarOption
from MountainCar import MountainCar
```

```python
options = [
    MountainCarOption(0, 1),
    MountainCarOption(1, 1),
    MountainCarOption(2, 1),

    MountainCarOption(0, 3),
    #MountainCarOption(1, 20),
    MountainCarOption(2, 3),

]

env = MountainCar()
mcts = MCTS(env, options)

action= mcts.run(5, 100)
#mcts.root.info(2)
```

```python
dot = graphviz.Digraph()
root = mcts.root
bfs_stack = [(0, root)]
while len(bfs_stack) > 0:
    action, node = bfs_stack.pop()
    
    dot.node(node.node_id, str((round(node.value), int(node.count))))
    
    if node.parent_node != None:
        dot.edge(node.parent_node.node_id, node.node_id, str(action))
        
    for action, child_node in node.child_nodes.items():
        bfs_stack.append((action, child_node))
```
