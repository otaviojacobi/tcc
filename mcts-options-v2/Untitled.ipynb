```python
with open('models/door_30_steps.pickle', 'rb') as f:
    m_door = pickle.load(f)
    
with open('models/random_30_steps.pickle', 'rb') as f:
    m_random = pickle.load(f)
    
with open('models/primitive_30_steps.pickle', 'rb') as f:
    m_prim = pickle.load(f)
```

```python
grid = -5000 * np.ones((13, 15))
```

```python
for k, v in m_prim['v'].items():
    grid[k.cur_x, k.cur_y] = v[0]
```

```python
_, ax = plt.subplots()
ax.imshow(grid)

# draw gridlines
ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
ax.set_xticks(np.arange(-0.5, 15, 1))
ax.set_yticks(np.arange(-0.5, 13, 1))

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.set_title('')

ax.plot()
```
