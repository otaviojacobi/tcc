```python
with open('models/door_30_steps.pickle', 'rb') as f:
    m_door = pickle.load(f)
    
with open('models/random_30_steps.pickle', 'rb') as f:
    m_random = pickle.load(f)
    
with open('models/primitive_30_steps.pickle', 'rb') as f:
    m_prim = pickle.load(f)
```

```python
grid = np.zeros((13, 15))

with open('easy.txt') as f:
    env_map = f.read()
    
env_map = env_map.split('\n')

for i in range(len(env_map)):
    for j in range(len(env_map[0])):
        if env_map[i][j] == '*':
            grid[i][j] = -1
        else:
            grid[i][j] = 0
            
            
model = m_random
for k, v in model['stats'].items():
    grid[k.cur_x, k.cur_y] =  v[1]#(v[0]//v[1]  - model['mini']) / (-model['mini']) #(v - model['mini']) / (-model['mini'])
```

```python
_, ax = plt.subplots()
ax.imshow(grid)

# draw gridlines
ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
ax.set_xticks(np.arange(-0.5, 15, 1))
ax.set_yticks(np.arange(-0.5, 13, 1))

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.set_title('')

ax.plot()
```
